```python
import pandas as pd
```

```python
#Load in csv
df = pd.read_csv('truliaOuput.csv')

#Get rid of unwanted columns
del df['Unnamed: 0']
```

```python
#Rename columns to better match description
df.columns = ['Price','SqFt','City','Address','Beds','Baths']
city_state = df['City'].to_list()

#Split up city column into city and state
for item in city_state:
    fixed_col = item.split(',')
    state = fixed_col[-1].strip()
    city = fixed_col[-2].strip()
    try:
        district = fixed_col[-3]
    except IndexError:
        district = 'N/A'
    
        
        
   
```

```python
#Add in new columns
df['City'] = city
df['State'] = state
#Re-arrange columns
df = df[['Address','City','State','Price','SqFt','Beds','Baths',]]
df
```

```python
#Export cleaned csv 
df.to_csv('Cleaned_Trulia.csv')
```
