```python
import pandas as pd
from prettypandas import PrettyPandas
```

```python
#Load in csv
df = pd.read_csv('zillowOuput.csv')
```

```python
#Get rid of unwanted columns
dfs = df.drop(['Unnamed: 0','typeProperty', 'yearBuilt', 'heating', 'cooling', 'parking', 'lot', 'saleHistory', 'taxHistory'],axis=1)
dfs
```

```python
#Renaming Columns to Match other CSV
dfs2 = dfs.rename({'Price (usd)': 'Price', 'Rooms': 'Beds', 'Bathrooms': 'Baths', 'Area (sqrt)': 'SqFt'}, axis=1)

#Seperating Address, City, and State into it's own seperate line
dfs2[["Address", "City", "State"]] = dfs2["Address"].str.split(pat=",", expand=True)

#Renaming
newdf = dfs2

#newdf["Zipcode"] = newdf["State"].str.split(pat=" ", expand=True)
#newdf

#Re-ordering the columns
orgdf = newdf[['Address', 'City', 'State', 'Price', 'SqFt', 'Beds', 'Baths', 'Link']]

#Changing Price to Currency format
PrettyPandas(orgdf).as_currency(subset='Price')
```
